# This example shows the plot of broken axis

```python
%matplotlib inline

# For the reference of the default color map of matplotlib, see
# https://matplotlib.org/users/dflt_style_changes.html

#For the fonts in matplotlib, please see
#https://matplotlib.org/users/dflt_style_changes.html#figure-size-font-size-and-screen-dpi
# This example is from CHAPTER 3 of the book "
# matplotlib for developpers",
# but I made many revisions

import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
mpl.rcParams['figure.facecolor'] = '1'
#if choose the grey backgroud, use 0.75
mpl.rcParams['figure.figsize'] = [8,6.5]
mpl.rcParams['lines.linewidth'] = 2.2
mpl.rcParams['legend.fancybox'] = True
mpl.rcParams["legend.framealpha"] = 0.7
mpl.rcParams['legend.fontsize'] = 19
mpl.rcParams['legend.scatterpoints'] = 1 #scatterpoints,
mpl.rcParams['legend.numpoints'] = 1
#it's the numer of maker points in the legend when
#creating a legend entry for a scatter plot
mpl.rcParams["axes.formatter.useoffset"]=False #turn off the axis offset-values. 
# If on. the axis label will use a offset value by the side of axis
mpl.rcParams["axes.linewidth"]  = 2.0 #change the boarder width
#plt.rcParams["axes.edgecolor"] = "0.15" #change the boarder color

ticklabel_size = 19
mpl.rcParams['xtick.labelsize'] = ticklabel_size
mpl.rcParams['ytick.labelsize'] = ticklabel_size

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame

fig1 = plt.figure()
fig1,(ax1_1, ax1_2) = plt.subplots(2, 1,sharex=True, facecolor='W', gridspec_kw={'height_ratios': [2, 1]})

x = np.linspace(-200,1000,100)
print(x)
y = x
ax1_1.plot(x,y)
ax1_2.plot(x,y)

# hide the spines between ax1 and ax2
ax1_1.spines['bottom'].set_visible(False)
ax1_2.spines['top'].set_visible(False)
# ax2.xaxis.tick_top()
ax1_2.tick_params(labeltop=False)
ax1_2.xaxis.tick_bottom()

d = .02 # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(transform=ax1_1.transAxes, color='k', clip_on=False)
ax1_1.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax1_1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax1_2.transAxes)  # switch to the bottom axes
ax1_2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax1_2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

## control the ticks
ax1_1.tick_params(direction='in', length=5, width=3.5, colors='k', which='major',
                grid_color='grey', grid_alpha=0.7, top=True, right=True, bottom=False)

ax1_2.tick_params(direction='in', length=5, width=3.5, colors='k', which='major',
                grid_color='grey', grid_alpha=0.7, top=False, right=True, bottom=True)

plt.xticks(rotation='30')

# ax.set_xlim(1e0, 1e3)
ax1_1.set_ylim(100, 800)
ax1_2.set_ylim(-200, -100)
plt.subplots_adjust(hspace=0.3)

ax1_1.set_ylabel('Y-axis',fontsize=22) 
ax1_2.set_xlabel('X-axis',fontsize=22) 


ax1_1.grid()
ax1_2.grid()
ax1_1.xaxis.grid(False)
ax1_2.xaxis.grid(False)

plt.tight_layout()
fig1.savefig("../images/broken-axis-ex01.pdf", bbox_inches='tight')
```
